### <div align = 'center'>**Newton's Method**</div>

---

Newton's Method (also called the **Newton-Raphson Method**) is an iterative numerical technique used to find the roots of a real-valued function \( f(x) \). It is widely used in solving nonlinear equations and optimization problems.

---

### **1. Mathematical Explanation**

#### **1.1 Formula**
The iterative formula for Newton's Method is:

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

Where:
- $ x_n $: Current approximation of the root.
- $ x_{n+1} $: Next approximation of the root.
- $ f(x) $: The function whose root is being calculated.
- $ f'(x) $: The derivative of $ f(x) $.

---

#### **1.2 Steps**
1. Start with an **initial guess** $ x_0 $.
2. Compute the next approximation $ x_{n+1} $ using the formula.
3. Repeat the process until the difference between successive approximations is smaller than a predefined tolerance (convergence).

---

#### **1.3 Convergence**
Newton's Method converges quadratically near the root if:
- The initial guess $ x_0 $ is close to the actual root.
- The derivative $ f'(x) $ is non-zero near the root.

---

### **2. Python Implementation**

```python
def newtons_method(f, f_prime, x0, tolerance=1e-6, max_iterations=100):
    """
    Newton's Method for finding roots of a function.
    
    Parameters:
        f: Function whose root is to be found.
        f_prime: Derivative of the function.
        x0: Initial guess.
        tolerance: Convergence tolerance.
        max_iterations: Maximum number of iterations.
    
    Returns:
        Approximation of the root.
    """
    x_n = x0
    for i in range(max_iterations):
        f_xn = f(x_n)
        f_prime_xn = f_prime(x_n)
        
        # Check if derivative is zero to avoid division by zero
        if f_prime_xn == 0:
            print("Derivative is zero. No solution found.")
            return None
        
        # Update the next approximation
        x_next = x_n - f_xn / f_prime_xn
        
        # Check for convergence
        if abs(x_next - x_n) < tolerance:
            print(f"Converged to {x_next} after {i+1} iterations.")
            return x_next
        
        x_n = x_next
    
    print("Maximum iterations reached. No solution found.")
    return None
```

### **3. Example Usage**

Letâ€™s solve $ f(x) = x^2 - 2 $ (i.e., find the square root of 2):



```python
# Define the function and its derivative
def f(x):
    return x**2 - 2

def f_prime(x):
    return 2 * x

# Initial guess
x0 = 1.0

# Call Newton's Method
root = newtons_method(f, f_prime, x0)

print(f"Root: {root}")
```



This is the square root of 2, calculated using Newton's Method.

### **4. Visualization**

To better understand how Newton's Method works, we can visualize the iterations:



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 - 2

def f_prime(x):
    return 2 * x

# Newton's Method Iterations
def newtons_method_visual(f, f_prime, x0, tolerance=1e-6, max_iterations=100):
    x_values = [x0]  # Store all approximations for visualization
    x_n = x0
    for i in range(max_iterations):
        f_xn = f(x_n)
        f_prime_xn = f_prime(x_n)
        
        if f_prime_xn == 0:
            print("Derivative is zero. No solution found.")
            return x_values
        
        x_next = x_n - f_xn / f_prime_xn
        x_values.append(x_next)
        
        if abs(x_next - x_n) < tolerance:
            break
        
        x_n = x_next
    
    return x_values

# Initial guess
x0 = 1.0

# Get all iterations
x_iterations = newtons_method_visual(f, f_prime, x0)

# Plot the function and iterations
x = np.linspace(0, 2, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = x^2 - 2", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
for i, x_val in enumerate(x_iterations):
    plt.scatter(x_val, f(x_val), color="red", label=f"Iteration {i}" if i == 0 else "")
plt.title("Newton's Method Iterations")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

### **5. Applications**

1. **Root Finding**:
   - Solving equations like $ f(x) = 0 $.

2. **Optimization**:
   - Finding critical points (minima or maxima) of a function.

3. **Machine Learning**:
   - Used in optimization algorithms like Newton's optimization method.

### **6. Summary**

Newton's Method is a powerful numerical technique for solving nonlinear equations. It converges quadratically near the root if the initial guess is close to the actual root. However, it requires the derivative of the function and is sensitive to the choice of the initial guess.
